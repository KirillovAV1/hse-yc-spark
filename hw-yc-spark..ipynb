Создание данных

```scala
%spark.pyspark
from pyspark.sql import functions as F
users = spark.createDataFrame(
    [
        ("u1", "Berlin"),
        ("u2", "Berlin"),
        ("u3", "Munich"),
        ("u4", "Hamburg"),
    ],
    ["user_id", "city"]
)
orders = spark.createDataFrame(
    [
        ("o1", "u1", "p1", 2, 10.0),
        ("o2", "u1", "p2", 1, 30.0),
        ("o3", "u2", "p1", 1, 10.0),
        ("o4", "u2", "p3", 5, 7.0),
        ("o5", "u3", "p2", 3, 30.0),
        ("o6", "u3", "p3", 1, 7.0),
        ("o7", "u4", "p1", 10, 10.0),
    ],
    ["order_id", "user_id", "product_id", "qty", "price"]
)
products = spark.createDataFrame(
    [
        ("p1", "Ring VOLA"),
        ("p2", "Ring POROG"),
        ("p3", "Ring TISHINA"),
    ],
    ["product_id", "product_name"]
)
users.show()
orders.show()
products.show()
```

1. Посчитать revenue = qty * price

```scala
%spark.pyspark

orders_with_revenue = orders.withColumn("revenue", F.col("qty") * F.col("price"))

orders_with_revenue.show()
```

2. Объединить orders с users и products

```scala
%spark.pyspark

tmp_orders = orders_with_revenue.join(
    F.broadcast(users),
    on="user_id",
    how="left"
)

tmp_orders = tmp_orders.join(
    F.broadcast(products),
    on="product_id",
    how="left"
)

tmp_orders.show()
```

3. Посчитать метрики по (city, product_id, product_name):
* orders_cnt (кол-во заказов)
* qty_sum (сумма qty)
* revenue_sum (сумма revenue)


```scala
%spark.pyspark

tmp_orders_agg = tmp_orders.groupBy("city", "product_id", "product_name").agg(
    F.count("order_id").alias("orders_cnt"),
    F.sum("qty").alias("qty_sum"),
    F.sum("revenue").alias("revenue_sum")
)

tmp_orders_agg.show()
```

4. Для каждого города выбрать Top-2 товара по revenue_sum используя Window


```scala
%spark.pyspark

from pyspark.sql.window import Window 
from pyspark.sql.functions import row_number, desc

w = Window.partitionBy("city").orderBy(desc("revenue_sum"))

mart_city_top_products = tmp_orders_agg.withColumn("rn", row_number().over(w)).filter(F.col("rn") <= 2)

mart_city_top_products.show()
```

5. Записать результат в s3 и в HDFS по пути :
/tmp/sandbox_zeppelin/mart_city_top_products/ (parquet, overwrite)

```scala
%spark.pyspark

path_s3 = "s3a://s3-etl/tmp/sandbox_zeppelin/mart_city_top_products/"
path_hdfs = "hdfs:///tmp/sandbox_zeppelin/mart_city_top_products/"

mart_city_top_products.write.mode("overwrite").parquet(path_s3)
mart_city_top_products.write.mode("overwrite").parquet(path_hdfs)
```

6. Прочитать обратно и показать топ-результат

```scala
%spark.pyspark

spark.read.parquet(path_s3).orderBy("city", "rn").show()
```
